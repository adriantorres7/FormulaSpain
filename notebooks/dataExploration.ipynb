```python
cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
tablas = cursor.fetchall()
print("Tablas:", tablas)

# Ver esquema y relaciones de cada tabla
for (tabla,) in tablas:
    print(f"\n--- {tabla} ---")
    cursor.execute(f"PRAGMA table_info({tabla});")
    columnas = cursor.fetchall()
    for col in columnas:
        print(f"Columna: {col[1]} - Tipo: {col[2]}")

    cursor.execute(f"PRAGMA foreign_key_list({tabla});")
    fks = cursor.fetchall()
    for fk in fks:
        print(f"🔗 FK: {fk[3]} → {fk[4]} en {fk[2]}")
```
